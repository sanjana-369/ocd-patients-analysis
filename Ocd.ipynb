```python
ethinicity_mapping = {'Caucasian':0, 'African':1, 'Asian':2, 'Hispanic':3}
df['Ethnicity'] = df['Ethnicity'].map(ethinicity_mapping)
print(df.head())
```

```python
gender_mapping={'Male':0, 'Female':1}

df['Gender']=df['Gender'].map(gender_mapping)

print(df.head())
```

```python
marital_status_mapping={'Single':0, 'Divorced':1, 'Married':2}

df['Marital Status']=df['Marital Status'].map(marital_status_mapping)

df.head()
```

```python
previous_diagnosis_mapping={'None':0, 'GDD':1, 'MDD':2, 'Pacic Disorder':3, 'PTSD':4}

df['Previous Diagnoses']=df['Previous Diagnoses'].map(previous_diagnosis_mapping)

df.head()
```

```python
family_history_mapping={'Yes':1, 'No':0}


df['Family History of OCD']=df['Family History of OCD'].map(family_history_mapping)

df.head()
```

```python
obsession_mapping={'Contamination':0, 'Ham-related':1, 'Hoarding':2, 'Religious':3, 'Symmentry':4}


df['Obsession Type']=df['Obsession Type'].map(obsession_mapping)

df.head()
```

```python
depression_diag = {'Yes':1, 'No':0}
anxiety_diag = {'Yes':1, 'No':0}
df['Depression Diagnosis'] = df['Depression Diagnosis'].map(depression_diag)
df['Anxiety Diagnosis'] = df['Anxiety Diagnosis'].map(anxiety_diag)
print(df.head())
```

```python
column_drop = [0,1,5,6,7,11,16]
df.drop(df.columns[column_drop], axis=1, inplace=True)
print(df.head())
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
Y_pred = classifier.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy Score : ',accuracy*100)
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
```

```python
true_positive = cm[0,0]
false_positive = cm[0,1]
true_negative = cm[1,1]
false_negative = cm[1,0]
sensitivity = true_positive / (true_positive+false_positive)
specificity = true_negative / (true_negative+false_negative)
LR_pos = sensitivity / (1-specificity)
LR_neg = (1-sensitivity)/specificity

print('Sensitivity of Dataset : ',sensitivity)
print('Specificity of Dataset : ',specificity)
print('LR+ : ',LR_pos)
print('LR- : ',LR_neg)
```
